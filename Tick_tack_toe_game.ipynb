```python
from IPython.display import clear_output
import random

#--------------------- FUNCTIONS -----------------#

# Print out a game board
def display_board(board):
    print('     |     |   ')
    print('  ' + board[7] + '  |  ' + board[8] + '  |  ' + board[9])
    print('     |     |   ')
    print('-----------------')
    print('     |     |   ')
    print('  ' + board[4] + '  |  ' + board[5] + '  |  ' + board[6])
    print('     |     |   ')
    print('-----------------')
    print('     |     |   ')
    print('  ' + board[1] + '  |  ' + board[2] + '  |  ' + board[3])
    print('     |     |   ')

# Player input to assign marker
def player_input():
    p1_marker=' '
    while (p1_marker.upper() not in 'XO'):
        p1_marker = input("Player1: Do you want to be X or O?  ")
    return p1_marker.upper() 

# Place marker at the desired position
def place_marker(board, marker, position):
    board[position]=marker.upper()

# Randomly choose which player goes first
def choose_first():
    player_chosen = random.randint(1,2)
    return player_chosen

# Check if the board is full
def full_board_check(board):
    if board[1] == ' ' or board[2] == ' ' or board[3] == ' ' or board[4] == ' ' or board[5] == ' ' or board[6] == ' ' or board[7] == ' ' or board[8] == ' ' or board[9] == ' ':
        return True
    else:
        return False
    
#  Check availability of a marker position
def space_check(board, position):
    if board[position] == ' ':
        return True
    else:
        return False    

# Choice of a position by a player to place the marker
def player_choice(board):
    while True:
        if  full_board_check(board):
            position = int(input('Choose your next position (1-9):  '))
            if space_check(board,position):
                return position
                break
            else:
                print('\nThis position is not available')
                position = int(input('Choose another position:  '))
                return position
                break
        else:
            print('\n The game is a draw')
            return False
            break
            
# To check if a marker has won
def win_check(board, mark):
    if (board[1] == board[2] == board[3] == mark or board[4] == board[5] == board[6] == mark or board[7] == board[8] == board[9] == mark or board[1] == board[4] == board[7] == mark or board[2] == board[5] == board[8] == mark or board[3] == board[6] == board[9] == mark or board[1] == board[5] == board[9] == mark or board[3] == board[5] == board[7] == mark):
        return True
    else:
        return False

# play again?
def replay():
    play=input('\nDo you want to play again? Enter Yes or No:  ')
    if play.lower()[0] == 'y':
        clear_output()
        return True
    else:
        return False
    
#----------------------- MAIN PROGRAM ----------------#
print('\n\nWelcome to Tic Tac Toe!')
print('\nEach position on the board is represented with a number as:\n')
num_pad=['#','1','2','3','4','5','6','7','8','9']
display_board(num_pad)

while True:
    player1_marker=player_input()
    
    if player1_marker == 'X':
        player2_marker = 'O'
    else:
        player2_marker = 'X'
        
    marker=['#',player1_marker,player2_marker]
    
    player_first = choose_first()
    print('\nPlayer '+ str(player_first) + ' will go first.')
    
    if player_first == 1:
        player_second = 2
    else:
        player_second = 1
    
    ready_to_play=input("\nAre you ready to play? Enter Yes or No.\n")
    clear_output()
    
    if ready_to_play.lower()[0] == 'y':
        game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
        display_board(game_board)
        
        while True:
            position=player_choice(game_board)
            if not position:
                break
            clear_output()
            place_marker(game_board, marker[player_first],position)
            display_board(game_board)
            if win_check(game_board, marker[player_first]):
                print('Congratulations! Player ' + str(player_first) + ' has won the game!')
                break
            
            position=player_choice(game_board)
            if not position:
                break
            place_marker(game_board,marker[player_second],position)
            clear_output()
            display_board(game_board) 
            if win_check(game_board,marker[player_second]):
                print('Congratulations Player ! ' + str(player_second) + ' has won the game!')
                break
    else:
        break
    
    if not replay():
        break   
```
